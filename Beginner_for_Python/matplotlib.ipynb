先週のグラフを表示するプログラムを示す．

```python
import numpy as np
import matplotlib.pyplot as plt

###############
# グラフ表示したい関数
###############
def mathfunc(x):
    y = x**4 - 8*x**3 + 18*x**2 - 3*x + 1
    return y


#######################
# 作成した関数のグラフを表示する
#######################

# 定義域を設定
xmin = -1.0
xmax = 4.5
num = 100 # xminからxmaxまでをnum個で区切る
x = np.linspace(xmin, xmax, num)

# 関数から値域を取得
y = mathfunc(x)

# 点どうしを直線でつなぐ
plt.plot(x, y)

# 適切な表示範囲を指定
plt.xlim(xmin, xmax)

# グリッド追加
plt.grid(True)

# 表示
plt.show()
```

## 散布図で遊ぼう

ランダムに値を作成して，画面上にプロットしてみましょう．

```python
import numpy as np
import matplotlib.pyplot as plt
 
# 乱数を生成
x = np.random.rand(100)
y = np.random.rand(100)
 
# 散布図を描画
plt.scatter(x, y)
```

散布図のスタイルをいろいろ変えてみましょう．

```python
import numpy as np
import matplotlib.pyplot as plt
 
x = np.random.rand(100)
y = np.random.rand(100)
 
plt.scatter(x, y, s=600, c="pink", alpha=0.5, linewidths="2", edgecolors="red")
```

カラーバーをつけてみましょう．

```python
import numpy as np
import matplotlib.pyplot as plt
 
x = np.random.rand(100)
y = np.random.rand(100)
 
# 乱数を 100 件生成
value = np.random.rand(100)

plt.scatter(x, y, s=100, c=value, cmap='Blues')

# カラーバーを表示
plt.colorbar()
```

もっと複雑なこともすぐできます！！

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x, y):
    return np.cos(x+y) ** 5 + np.cos(4 + y * x) * np.cos(-x)

x = np.linspace(0, 3, 100)
y = np.linspace(0, 3, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig, ax = plt.subplots()
ax.contour(X, Y, Z, colors='g')
ax.set_aspect('equal','box')
plt.savefig("contour.png", bbox_inches = 'tight', pad_inches = 0.1)
```

```python
fig, ax = plt.subplots()
cont10 = ax.contour(X, Y, Z, 10, cmap='PuOr')
ax.set_aspect('equal','box')
plt.colorbar(cont10)
plt.savefig("contour_PuOr.png", bbox_inches = 'tight', pad_inches = 0.1)
```

```python
fig, ax = plt.subplots()
contf = ax.contourf(X, Y, Z, 10, cmap='PuOr')
ax.set_aspect('equal','box')
plt.colorbar(contf)
plt.savefig("contourf_PuOr.png", bbox_inches = 'tight', pad_inches = 0.1)
```
